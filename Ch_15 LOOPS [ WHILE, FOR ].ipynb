```python
# WHILE LOOP

i = 1

while i <= 10:
    print(i)
    i += 1
print("Done with the Loop")
```

```python
# BUILDING A GUESSING GAME USING WHILE LOOP

secret_word = "BTS"
guess = ""

while guess != secret_word:
    guess = input("Enter Guess: ")
    
print("You win!")

print("Now we are giving the limit to the user to guess the word")

secret_word = "bts"
guess = ""
guess_count = 0
guess_limit = 3
out_of_guesses = False  

while guess != secret_word and not(out_of_guesses):
    if guess_count < guess_limit:
        guess = input("Enter the guess: ")
        guess_count += 1
    else:
        out_of_guesses = True

if out_of_guesses:
    print(" Out of Guesses, you LOSE!")
else:
    print("You WIN!")
```

```python
# 2D LISTS & NESTED LOOPS

number_grid = [
    [1,2,3],
    [4,5,6],
    [7,8,9],
    [0]
]

print(number_grid)
print(number_grid[0])
print(number_grid[0][0])
print(number_grid[2][1])

print("")


for row in number_grid:
    print(row)

print("")

for row in number_grid:
    for col in row:
        print(col)
```

```python
# BUILD A TRANSLATOR

def translate(phrase):
    translation = ""
    for letter in phrase:
        if letter in "AEIOUaeiou":
            translation = translation + "g"
        else:
            translation = translation + letter
    return translation

print(translate(input("Enter a phrase: ")))

print("")
print("Printing with the lower case 'g' ")
def translate(phrase):
    translation = ""
    for letter in phrase:
        if letter.lower() in "aeiou":
            translation = translation + "g"
        else:
            translation = translation + letter
    return translation

print(translate(input("Enter a phrase: ")))

print("")
print("Printing with the UPPER case 'G' ")
def translate(phrase):
    translation = ""
    for letter in phrase:
        if letter.lower() in "aeiou":
            if letter.isupper():
                translation = translation + "G"
            else:
                translation = translation + "g"
        else:
            translation = translation + letter
    return translation

print(translate(input("Enter a phrase: ")))
```
