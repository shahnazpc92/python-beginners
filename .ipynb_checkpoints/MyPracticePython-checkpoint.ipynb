```python
# IF STATEMENT & COMPARISONS

def max_num(n1, n2, n3):
    if n1 >= n2 and n1 >= n3:
        return n1
    elif n2 >= n1 and n2 >= n3:
        return n2
    else:
        return n3
    
max_num(50,20,8)   # We can compare strings also and other operaters like !=,==,>,< etc.
```

```python
# BUILDING A GUESSING GAME

secret_word = "BTS"
guess = ""

while guess != secret_word:
    guess = input("Enter Guess: ")
    
print("You win!")

print("Now we are giving the limit to the user to guess the word")

secret_word = "bts"
guess = ""
guess_count = 0
guess_limit = 3
out_of_guesses = False  

while guess != secret_word and not(out_of_guesses):
    if guess_count < guess_limit:
        guess = input("Enter the guess: ")
        guess_count += 1
    else:
        out_of_guesses = True

if out_of_guesses:
    print(" Out of Guesses, you LOSE!")
else:
    print("You WIN!")
```

```python
# EXPONENT FUNCTION

def raise_to_power(base_num,pow_num):
    result = 1
    for index in range(pow_num):
        result = result * base_num
    return result
print(raise_to_power(2,3))
```

```python
# BUILD A TRANSLATOR

def translate(phrase):
    translation = ""
    for letter in phrase:
        if letter in "AEIOUaeiou":
            translation = translation + "g"
        else:
            translation = translation + letter
    return translation

print(translate(input("Enter a phrase: ")))

print("")
print("Printing with the lower case 'g' ")
def translate(phrase):
    translation = ""
    for letter in phrase:
        if letter.lower() in "aeiou":
            translation = translation + "g"
        else:
            translation = translation + letter
    return translation

print(translate(input("Enter a phrase: ")))

print("")
print("Printing with the UPPER case 'G' ")
def translate(phrase):
    translation = ""
    for letter in phrase:
        if letter.lower() in "aeiou":
            if letter.isupper():
                translation = translation + "G"
            else:
                translation = translation + "g"
        else:
            translation = translation + letter
    return translation

print(translate(input("Enter a phrase: ")))
```
