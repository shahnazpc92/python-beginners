```python
# LISTS FUNCTIONS

friends = ["karen", "mike", "nancy", "tomy", "dom"]
print("List is printing:-",friends)
print("")

lucky_numbers = [2,1,3,5,4,7]
friends = ["karen", "mike", "nancy", "tomy", "dom"]
friends.extend(lucky_numbers)
print("Extend function add the lucky_number list to the friends list:-",friends)
print("")

friends = ["karen", "mike", "nancy", "tomy", "dom"]
friends.append("kelly")
print("Append function add the name at the end of the friend list:-",friends)
print("")

friends = ["karen", "mike", "nancy", "tomy", "dom"]
friends.insert(1,"loki")
print("Insert function insert the name at the indexed value which we have given:-",friends)
print("")

friends = ["karen", "mike", "nancy", "tomy", "dom"]
friends.remove("nancy")
print("Remove function remove the name from the list:-",friends)
print("")

friends = ["karen", "mike", "nancy", "tomy", "dom"]
friends.clear()
print("Clear function clear or give empty set of the list:-",friends)
print("")

friends = ["karen", "mike", "nancy", "tomy", "dom"]
friends.pop()
print("Pop function eleminate the last element of the list:-",friends)
print("")

friends = ["karen", "mike", "nancy", "tomy", "dom"]
print("Index function gives the index value of the element:-",friends.index("tomy")) #Index funct should written in print
print("")

friends = ["karen", "mike", "nancy", "tomy", "tomy", "tomy", "tomy", "dom"]
print("Count function count the value of the element in the list:-",friends.count("tomy")) #Count funct should written in print
print("")

friends = ["karen", "mike", "nancy", "tomy", "tomy", "tomy", "tomy", "dom"]
friends.sort()
print("Sort function sort into accending or alphabetical order to the list:-",friends) #Sort funct should not written in print
print("")

lucky_numbers = [5,4,7,2,1,3]
lucky_numbers.sort()
print("Sort function sort into accending order to the list:-",lucky_numbers) #Sort funct should'nt written in print
print("")

lucky_numbers = [5,4,7,2,1,3]
lucky_numbers.reverse()
print("Reverse function reverse the list:-",lucky_numbers) 
print("")

friends = ["karen", "mike", "nancy", "tomy", "tomy", "tomy", "tomy", "dom"]
friends2 = friends.copy()
print("Copy function copy the whole list of friends int0 friends2:-",friends2) 
print("")
```
