```python
# IF STATEMENT & COMPARISONS

def max_num(n1, n2, n3):
    if n1 >= n2 and n1 >= n3:
        return n1
    elif n2 >= n1 and n2 >= n3:
        return n2
    else:
        return n3
    
max_num(50,20,8)   # We can compare strings also and other operaters like !=,==,>,< etc.
```
